```python
cp_data = data.copy()
cp_data
```

```python
# Checking out the unique dates of FitBit Dataset
cp_data['ActivityDate'].unique()
```

```python
# adding the yearm month and date columns to the dataset

cp_data['year'] = pd.DatetimeIndex(cp_data['ActivityDate']).year
cp_data['month'] = pd.DatetimeIndex(cp_data['ActivityDate']).month
cp_data['date'] = pd.DatetimeIndex(cp_data['ActivityDate']).day
cp_data['Day_Name'] = pd.DatetimeIndex(cp_data['ActivityDate']).day_name()
```

```python
cp_data.head(10)
```

```python
### Groupby the day of the month and make a boxplot of calories burnt

plt.figure(figsize=(15,8))
box = sns.boxplot(x='date', y='Calories', data=cp_data)
box = sns.swarmplot(x='date', y='Calories', data=cp_data, color="grey")

box.set_title('Box plot of Calories with day of the month')
```

```python
plt.figure(figsize=(10,5))
br = sns.barplot(x='Day_Name', y='Calories',  data=cp_data)
br.set_title('Barplot of calories by the day of the week')
```

```python
sns.relplot(data=cp_data, y='Calories', x='TotalSteps', hue='TotalDistance')
```

```python
sns.pairplot(data=cp_data[['TotalSteps', 'TotalDistance']])
```

```python
sns.relplot(x='date', y='Calories', data=cp_data, hue='Day_Name')
```

```python
cp_data.plot(kind='box', x='Calories', y='TotalSteps')
```

```python
plt.figure(figsize=(15,8))
fg = sns.scatterplot(x='Calories', y='LightlyActiveMinutes', data=cp_data)

fg.set_title('Scatterplot of calories and intense_activities')
```

```python
plt.figure(figsize=(15,8))
fg = sns.scatterplot(x='Calories', y='TotalDistance', data=cp_data)
fg.set_title('Scatterplot of calories and intense_activities')
```

```python
set = ['TotalDistance','LoggedActivitiesDistance','VeryActiveDistance','ModeratelyActiveDistance', 'LightActiveDistance']
wide_df= cp_data[set]

plt.figure(figsize=(15,8))
fg = sns.lineplot(data=wide_df)
fg.set_title('Un-normalized value of calories and different activities based on distance')
```

##### The EDA here gives that the insight about the relation between the active hours, the distance for which the user has moderate and intense activity and the calories burnt during that period.
